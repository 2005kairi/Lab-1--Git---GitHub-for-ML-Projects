```python
import pandas as pd

# full file path for the dataset
file_path = 'C:/Users/Luvje/Downloads/archive/emails.csv'

# Load the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)

# Function to shorten text to first 10 words
def shorten_text(text):
    words = str(text).split()
    return " ".join(words[:10]) + "..." if len(words) > 10 else text

# Apply shortening only for display
df_display = df.copy()
df_display['text'] = df_display['text'].apply(shorten_text)

# Style: bold + leftâ€‘aligned headers, normal cells
styled_df = df_display.head().style.set_table_styles([
    {'selector': 'th.col_heading', 'props': [('font-weight', 'bold'),
                                             ('text-align', 'left')]},
    {'selector': 'td', 'props': [('text-align', 'left')]}
])

print(f"Shape of the dataset: {df.shape}")
styled_df
```
